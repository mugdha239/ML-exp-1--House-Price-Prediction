```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("housing.csv")
```

```python
data
```

```python
data.info()
```

```python
data.dropna(inplace=True)
```

```python
data.info()
```

```python
from sklearn.model_selection import train_test_split

x = data.drop(['median_house_value'],axis=1) #dropping the target class column 
y = data['median_house_value'] #target class column stored here
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2) #0.2 is the % of data we are using for testing..here 20%
```

```python
train_data = x_train.join(y_train)
```

```python
train_data.corr()
# correlation among attributes  ... each attribute has a correlation of 1 with itself
# helpful to see the correlation of attributes with our target class
```

```python
plt.figure(figsize=(15,8))
sns.heatmap(train_data.corr(),annot=True,cmap="YlGnBu")
# Exploration .... color gradation is used to show correlation 
# Four our target class i.e. median_house_value , the attribute median_income correlates highly and thus the gradation is much darker
```

```python
train_data = train_data.join(pd.get_dummies(train_data.ocean_proximity)).drop(['ocean_proximity'],axis=1)
#joining the dummy value columns to our train_data 
#Only that value is 1 which is the proximity from the ocean ; others are 0
#Now we have converted the values into binary for ocean proximity , we can drop the column which holds the actual textual values of ocean proximity
```

```python
test_data = x_test.join(y_test)
```

```python
test_data.ocean_proximity.value_counts()
```

```python
test_data = test_data.join(pd.get_dummies(test_data.ocean_proximity)).drop(['ocean_proximity'], axis =1)
```

```python
test_data
```

```python
from sklearn.linear_model import LinearRegression

x_train,y_train = train_data.drop(['median_house_value'],axis = 1),train_data['median_house_value']

reg = LinearRegression()

reg.fit(x_train, y_train)
```

```python
x_test, y_test = test_data.drop(['median_house_value'],axis = 1),test_data['median_house_value']
```

```python
reg.score(x_test,y_test)
```
